### (1)import必要的套件

```python
#數據、矩陣處理套件numpy
import numpy as np
#繪圖處理套件matplotlib
import matplotlib.pyplot as plt
#繪圖處理套件顯示中文matplotlib.font_manager
import matplotlib.font_manager as plt_font
#設定中文字體物件和字型檔案路徑
twfont1 = plt_font.FontProperties(fname="drive/My Drive/colab/kaiu.ttf")
```

### (2)定義多項式函數f(x)=x<sup>4</sup>-38x<sup>2</sup>-48x+24

```python
def F(x):
    #回傳函數的運算結果
    return x**4-38*x**2-48*x+24
```

### (3)定義原函數的導函數f'(x)

```python
def F1(x):
    #回傳導函數的運算結果
    return  4*x**3-76*x-48   
```

### (4)畫出原函數F(x)和F<sub>1</sub>(x)的導函數，y軸為零的水平線

```python
#設定繪圖區大小
plt.figure(figsize=(6,3))
#繪圖區的標題，設定用中文字體twfont1，字體大小20
plt.title("利用導函數找函數的極值",fontproperties=twfont1,fontsize=20)
#設定橫軸和縱軸的標題 
plt.xlabel("x",fontsize=20) 
plt.ylabel("y",fontsize=20)
#產生橫軸座標的數據點矩陣(在-7和7之間取1000個點的數據，含-7含和7)
x = np.linspace(-7,7, 1000)
#畫圖plot(橫軸數據點矩陣，對應橫軸的含數值矩陣，繪圖線的label)
#劃出原函數的關係圖
plt.plot(x,F(x),label="原函數")
#劃出導函數的關係圖
plt.plot(x,F1(x),label="導函數")
#劃出縱軸為零的橫線
plt.plot(x,x*0,label="y=0")
#顯示繪圖線的label圖示
plt.legend(prop=twfont1)
```
