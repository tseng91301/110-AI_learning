### (1)import必要的套件

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as plt_font
twfont1=plt_font.FontProperties(fname="drive/My Drive/colab/kaiu.ttf")
```

### (2)定義二變數函數f(x,y)=x<sup>2</sup>+y<sup>2</sup>




```python
def F(x,y):
    return x**2+y**2
```

### (3)鋪設繪圖的底圖網格

```python
#產生橫軸、縱軸座標的數據點矩陣(在-5和5之間，間隔0.1)
x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
#利用橫軸、縱軸座標的數據點鋪設網點
x,y=np.meshgrid(x,y)
```

```python
print(x,y,sep="\n")
```

### (4)劃出曲面圖

```python
#開啟繪圖區域
fig=plt.figure(figsize=(12,9))
#設定為 3D 繪圖
ax=fig.gca(projection='3d')
#設定z軸範圍
ax.set_zlim(0,50)
#設定坐標軸名稱
ax.set_xlabel('X軸',fontproperties=twfont1)
ax.set_ylabel('Y軸',fontproperties=twfont1)
ax.set_zlabel('Z軸',fontproperties=twfont1)
#畫出3D曲面,cmap為數值及色階對應方式,antiliased=Fasle可以加快繪圖速度 
surf=ax.plot_surface(x, y, F(x,y), cmap="cool",alpha=0.2, antialiased=False)
#顯示數值及色階對應方式,shrink設定colorbar長度為圖片高度的幾倍,預設值1.0,aspect設定colorbar寬、高比,預設值20
fig.colorbar(surf,shrink=0.5,aspect=5)
#elev=20設定視線和水平面的夾20度，azim=30設定立體圖順時針旋轉30度
ax.view_init(elev=30,azim=30)
```

### (5)劃出等高線

```python
# 開啟繪圖區域
plt.figure(figsize=(5,5))
plt.title('等高線圖',fontproperties=twfont1,fontsize=20)
CS=plt.contour(x,y,F(x,y),10)
plt.clabel(CS, inline=1, fontsize=10)
```
